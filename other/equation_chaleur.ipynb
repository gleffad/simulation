```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import solve
from scipy import misc
import sympy as sp
```

```python

def t(x):
    return np.exp(-10*(x-0.3)**2)

def equa_chaleur(Tg,Td,Nx,Nt,T,L):

    dt = T/Nt
    dx = L/Nx

    duree = np.linspace(0,T,Nt)
    X = np.linspace(0,L,Nx)
    T = np.zeros((Nt+1,Nx+1))

    alpha = dt/(dx**2)

    for i in range(Nx+1):
        T[0][i] = t(i*dx)
    
    for n in range(1,Nt+1):
        T[n][0] = Tg

    for n in range(1,Nt+1):
        T[n][Nx] = Td
    
    for n in range(Nt):
        for i in range(1,Nx):
            T[n+1][i] = T[n][i] + alpha *(T[n][i+1] - 2*T[n][i] + T[n][i-1])

    return T

L = 1
T = 1
Tg = 0.4
Td = 0
Nx = 50
Nt = 10000

sol = equa_chaleur(Tg,Td,Nx,Nt,T,L)

i = 0
while i < 1000 : 
    plt.plot(sol[i,:],"-r")
    i = i + 125
```
