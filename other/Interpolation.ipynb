```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import solve
```

# Interpolation avec 2 points

```python
X=[0,1]
Y=[2,5]
plt.plot(X,Y,"b*")

A = np.array([[1,X[0]],[1,X[1]]])
b = np.array([Y[0],Y[1]])
sol = solve(A,b)
def P1(x): return sol[0] + sol[1]*x

# Methode directe avec les polynomes de Lagrange
def P1bis(x): return Y[1]*(x-X[0])/(X[1]-X[0])+Y[0]*(x-X[1])/(X[0]-X[1])

Vx=np.linspace(0,1,100)
plt.plot(Vx,P1(Vx),"-r",linewidth=3.5)
```

# Interpolation avec 3 points

```python
X=[0,1,2]
Y=[-2,5,6]
plt.plot(X,Y,"b*")

A = np.array([[1,X[0],X[0]**2],[1,X[1],X[1]**2],[1,X[2],X[2]**2]])
b = np.array([Y[0],Y[1],Y[2]])
sol = solve(A,b)
def P2(x): return sol[0] + sol[1]*x + sol[2]*x**2

# Methode directe avec les polynomes de Lagrange
def P2bis(x): return Y[0]*(x-X[1])*(x-X[2])/((X[0]-X[1])*(X[0]-X[2])) + Y[1]*(x-X[0])*(x-X[2])/((X[1]-X[0])*(X[1]-X[2]))\
                    + Y[2]*(x-X[0])*(x-X[1])/((X[2]-X[0])*(X[2]-X[1]))

Vx=np.linspace(0,2,100)
plt.plot(Vx,P2(Vx),"-r",linewidth=3.5)
```

# Interpolation , cas général

```python
X = [-2, -1, 0, 1, 2, 3, 4, 5]
Y = [0.1, 0.2, 0.5, 1, 0.5, 0.2, 0.1, 0.0588]

plt.plot(X,Y,"*b")

n = len(X)
A = np.array([ [X[i]**j for j in range(n)] for i in range(n) ])
Sol = solve(A,Y)
def Pn(x): return np.sum([Sol[i]*x**i for i in range(n)])

# Methode directe avec les polynomes de Lagrange
def Pnbis(x): return np.sum([Y[i]*np.product([(x - X[j])/(X[i] - X[j]) for j in range(n) if j != i]) for i in range(n)])

Vx = np.linspace(-2,5,200)
plt.plot(Vx,[Pn(x) for x in Vx],"-r",linewidth=5)


```
